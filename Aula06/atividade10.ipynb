```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('car_prices.csv')
data = df.copy()
```

```python
data
```

```python
data = data.dropna()
```

```python
data.describe()
```

```python
data_a= data.drop(["trim","vin","state","condition","odometer","interior","seller","mmr"],axis=1)
```

```python
data_a
```

```python
data_filter = data_a.loc[lambda table:table['year']==2010]
```

```python
# Contando os valores de 'make'
data_counts = data_filter['make'].value_counts().reset_index() 
```

```python
data_counts
```

```python
top_car= data_counts.sort_values('count',ascending=False).head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

f, ax = plt.subplots(figsize=(20,5))

ax.bar_label(ax.bar(top_car["make"],top_car["count"]),padding=0.3)
sns.barplot(data= top_car, x=top_car["make"], y=top_car["count"], palette="bright")
ax.set_title("Top 5 Sales Car in 10s per Make")
ax.set_xlabel("Make")
ax.set_ylabel("Number of Sales")
ax.set_ylim(0, top_car["count"].max() * 1.1)
plt.show()
```

```python
## converter valor para maiusculo
data_a['body'] = [value.upper() for value in data_a['body']]
```

```python
data_a
```

```python
first_make =data_counts['make'][0]
```

```python
data_filter_body = data_a.loc[(data_a['year']==2010)&(data_a['make'] == first_make)]
```

```python
data_counts_body = data_filter_body['body'].value_counts().reset_index()
```

```python
data_counts_body
```

```python
top_body = data_counts_body.sort_values('count',ascending=False).head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

f, ax = plt.subplots(figsize=(20,5))

ax.bar_label(ax.bar(top_body["body"],top_body["count"]),padding=0.3)
sns.barplot(data= top_body, x=top_body["body"], y=top_body["count"], palette="bright")
ax.set_title(f'Top 3 Sales Car per Body in 10s by {first_make}')
ax.set_xlabel("Body")
ax.set_ylabel("Number of Sales Car with that Body")
ax.set_ylim(0, top_body["count"].max() * 1.1)
plt.show()
```
